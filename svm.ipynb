```python
import numpy as np
```

```python
class svm_classifier():

  # initiating the hyper parameters
  def __init__(self,learning_rate,no_of_iterations,lambda_parameter):
    self.learning_rate=learning_rate
    self.no_of_iterations=no_of_iterations
    self.lambda_parameter=lambda_parameter

  # fitting the dataset to svm classifier
  def fit(self,x,y):
    m,n = x.shape
    self.m=m # no of datapoints or no of rows
    self.n=n # no of ip features or no of columns

    self.w=np.zeros(self.n)
    self.b=0

    self.x=x
    self.y=y

    # implementing gradient descent algorithm for optimization

    for i in range(self.no_of_iterations):
      self.update_weights()

  #function for updating weight and bias values
  def update_weights(self):
    # label encoding
    y_label=np.where(self.y <=0 ,-1,1)

    for index, x_i in enumerate(self.x):
      condition = y_label[index] * (np.dot(x_i,self.w)-self.b)>=1
      # index gives the zero based index from which we extract y_label
      # x_i is the iterator which gives the values of self.x one by one

      if(condition == True):

        dw=2*self.lambda_parameter*self.w
        db=0

      else:
        dw= 2*self.lambda_parameter*self.w - np.dot(x_i, y_label[index])
        db= y_label[index]

      self.w= self.w - self.learning_rate* dw
      self.b = self.b - self.learning_rate * db


  # predict the label for the given input values
  def predict(self,x):
    output = np.dot(x,self.w) - self.b
    predicted_labels = np.sign(output) # checking sign of op and classifying it based on sign
    y_hat = np.where(predicted_labels<=-1 ,0,1)
    # 0 represents person wihtout diabetes and 1 is  with diabetes

    return y_hat
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Data collection and pre processing

# loading dataset to pandas dataframe

diabetes_df = pd.read_csv('/content/diabetes.csv')
```

```python
# separating the features and targets
features = diabetes_df.drop(columns='Outcome',axis=1)
target = diabetes_df['Outcome']
```

```python
# standardising the data
scaler = StandardScaler()
```

```python
scaler.fit(features)
```

```python
standardised_data=scaler.transform(features)
```

```python
print(standardised_data)
```

```python
features = standardised_data # ranaming standardised data as fatures
# for our convinience
```

```python
# splitting data into train and test
x_train,x_test,y_train,y_test  = train_test_split(features,target,test_size=0.2,random_state=2)
```

```python
print(features.shape, x_train.shape, x_test.shape)
```

```python
# Training the svm_classifier model

classifier = svm_classifier(learning_rate=0.001,no_of_iterations=1000,lambda_parameter=0.01)
```

```python
# training svm classifier with training data
classifier.fit(x_train,y_train)
```

```python
# testing our model
# Finding accuracy
x_train_pred = classifier.predict(x_train)

training_data_accuracy = accuracy_score(y_train,x_train_pred)
print(training_data_accuracy)
```

```python
# accuracy on test data
x_test_pred = classifier.predict(x_test)

testing_data_accuracy=accuracy_score(y_test,x_test_pred)
print(testing_data_accuracy)
```
